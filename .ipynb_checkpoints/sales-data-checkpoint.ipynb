Question 2: What are the total sales attributed to customers over the age of 45?
Answer: 82039768.15

```python
df2.rename(columns = {"id":"customer_id"}, inplace = True)
```

```python
df3 = pd.merge(df, df2, on="customer_id", how="inner")
```

```python
df3.query("age > 45")["total_sales"].sum()
```

Question 3: How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)

```python
df3["invoice_date"].value_counts().sort_index()
```

```python
df3['invoice_date'] = pd.to_datetime(df3['invoice_date'])
daily_counts = df3['invoice_date'].value_counts().sort_index()
monthly_counts = df3.groupby(df3['invoice_date'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
plt.plot(monthly_counts.index.astype(str), monthly_counts.values, marker='o', linestyle='-')
plt.title('Total Count of Invoice Dates by Month')
plt.xlabel('Invoice Date (Month)')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Hard

Question 1: Create a pivot table showing the breakdown of sales across these dimensions, in this order: category, decade age range (e.g., 10-19, 20-29, and so forth).

```python
age_bins = [10, 19, 29, 39, 49, 59, 69, 79, 89, 99]
age_labels = ["10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80-89", "90-99"]
df3['age_range'] = pd.cut(df3['age'], bins=age_bins, labels=age_labels)

pivot_table = pd.pivot_table(df3, values='total_sales', index='category', columns='age_range', aggfunc='sum', fill_value=0)
print(pivot_table)
```
